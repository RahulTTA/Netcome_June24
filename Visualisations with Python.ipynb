```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.title('Example Line Plot')
plt.show()
```

```python
x = [1, 2, 3, 4, 5,6]
y = [2, 4, 6, 8, 10,40]

plt.scatter(x, y)
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.title('Example Scatter Plot')
plt.show()
```

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [2, 4, 6, 8, 10]

plt.bar(x, y)
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.title('Example Bar Plot')
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [15, 30, 45, 10, 5]

plt.pie(sizes, labels=labels)
plt.title('Example Pie Chart')
plt.show()
```

```python
import numpy as np

x = np.random.randn(1000)

plt.hist(x, bins=10)
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.title('Example Histogram')
plt.show()
```

```python
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,                      # DataFrame containing the data
    x="total_bill", y="tip",        # Specify x and y axis variables
    col="time",                     # Separate plots by the 'time' column
    hue="smoker",                   # Colour points based on the 'smoker' column
    style="smoker",                 # Differentiate markers by 'smoker'
    size="size"                     # Vary marker size based on 'size' column
)

```

```python
# Load the 'flights' dataset from Seaborn
flights = sns.load_dataset('flights')

# Pivot the dataset to reshape it for heatmap plotting
# Index: 'month', Columns: 'year', Values: 'passengers'
flights_pivoted = flights.pivot(index='month', columns='year', values='passengers')

# Create a heatmap visualization using the pivoted dataset
sns.heatmap(flights_pivoted, cmap='coolwarm')

```
